```python
import pandas as pd
```

```python
# Save file url into variable
file_url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data'
# Save data as a DataFrame
df = pd.read_csv(file_url, header=None)
df.head()
```

## 1. Rename Columns

> Rename 0 : 'Flow_ID', 1 : 'Date', 2 : 'Time', 3 : 'Count'

```python
# Rename columns using dictionary substitution
df = df.rename(columns={0:'Flow_ID',
                        1:'Date',
                        2:'Time',
                        3:'Count'})
df.head()
```

## 2. Selecting Data



1.   Select all data for the date July 24 2005 having flow id = 7
2.   From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10



```python
# Create a filter for the date of July 24 2005
date_filter = df['Date'] == '07/24/05'
# Create a filter for flow id's = 7
id_filter = df['Flow_ID'] == 7

# Save new, filtered dataframe with all data for the date of July 24, 2005 and a flow id of 7
df_filtered = df[date_filter & id_filter]
len(df_filtered)
```

```python
# Sort the filtered dataframe on 'Count', in descending order
df_filtered = df_filtered.sort_values(by='Count', ascending = False)
# Save top 10 rows to a variable called 'top10'
top10 = df_filtered.head(10)
top10
```

## 3. Subtracting from a Column

For the 10 rows outputted above, use pandas to subtract the lowest value of the 10 from all of them and then output the average value using the resulting counts. 

```python
# Use .min() function to determine the minimum value in the count column
counts_min = top10['Count'].min()
# Subtract that minimum value from all values in the count column
top10['Count'] - counts_min;
# display top10 to verify the count column is less by the value of the minimum
top10
```

```python
# Output average value of the resulting count column
top10['Count'].mean()
```

## 4. Grouping



1.   Select data in the month of August 2005 having flow id = 7
2.   Group the data based on date and get the max count per date



```python
# Create filter for the month of August
month_filter = df['Date'].str.startswith('08')
# Create filter to ensure the year is 2005
year_filter = df['Date'].str.endswith('05')

# Create new, filtered dataframe for data in August 2005, having flow id = 7
df_filtered2 = df[month_filter & year_filter & id_filter]
df_filtered2
```

```python
# Group the data based on date and get the max count per date
df_filtered2.groupby('Date')['Count'].max()
```

```python
df_filtered2.groupby('Date').max()
```
